```python
class ImportData(BaseModel):
  """this class manages the task of importing data from UCI ML Repository and loading data into a dataframe"""

  repoId: Optional[int] = Field(None, description="Repository ID for our dataset")
  data: Optional[Any] = None

  def grabData(self):
    """from given repId, this function downloads the data"""
    if not self.repoId:
      print(f"Please instantiate repo Id")
    else:
      try:
        self.data = fetch_ucirepo(id=self.repoId)
        return self.data
      except Exception as e:
        print(f'there is an error in fetching the data. Error code: {e}')
        return None
```

# Data Cleaning

### Removing classes 3 and 9 in Y

From the EDA, we learn that classes 3 and 9 have barely any responses in our dataset, therefore, there is no power in training for these classes. They should be removed from the dataset.

```python
Y2 = Y[(Y['quality'] != 3) & (Y['quality'] != 9)]
X2 = X[(Y['quality'] != 3) & (Y['quality'] != 9)]
```

### Outlier detection and removal

From EDA, we learn that there are significant outliers in our dataset, particularly for features fixed_acidity, citric_acid, free_sulfur_dioxide. We use descriptive statistics to guide us on the threshold to remove such datapoints

```python
Y2 = Y2[(X2['fixed_acidity'] < 12) &
               (X2['citric_acid'] < 0.6) &
               (X2['free_sulfur_dioxide'] < 100)]
X2 = X2[(X2['fixed_acidity'] < 12) &
               (X2['citric_acid'] < 0.6) &
               (X2['free_sulfur_dioxide'] < 100)]
```

We removed about 230 data points from our dataset (about 3%)

### Feature Engineering: using unsupervised method to create labels for White Wine and Red Wine

In the paper, the author mentioned that the samples contained 1599 red wines and 4898 white wines.

*  However, as you see from the given X and Y attributes above, we only have physiochemical data of the wine and the author did not release the labels for red and white wine.
*   The paper also stated that red and white wine have distinct physiochemical properties. Here I use clustering methods on our datasets to derive a proxy for red and white wines.
*   I find that using K-means clustering on first principal component of our feature results in two clusters of dataset that have very similar physiochemical data characteristics for the red and white wines in the paper. Therefore, I will use these sample clusters as proxies for Red and White wines.













```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import itertools
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
X_standard = pd.DataFrame(StandardScaler().fit_transform(X2)) # Standardization to do PCA

pca = PCA(n_components=1) # PCA with 1 principal component
X_pca = pca.fit_transform(X_standard)

kmeans = KMeans(n_clusters=2) # K-means with 2 clusters
kmeans.fit(X_pca) # Use K-means on X projected upon first principal component

clusters = kmeans.labels_ # obtain cluster labels on data

# include cluster label to our dataset
X2['cluster'] = clusters
X_standard['cluster'] = clusters
Y2['cluster'] = clusters
columns = X2.columns[:-1]
combinations = itertools.combinations(columns, 3)

# plot separation of cluster across various features in X
fig = plt.figure(figsize=(15, 10))
for i, (col1, col2, col3) in enumerate(combinations, 1):
    ax = fig.add_subplot(3, 3, i, projection='3d')
    ax.scatter(X2[col1], X2[col2], X2[col3], c=X2['cluster'], cmap='viridis', edgecolor='k')
    ax.set_xlabel(col1)
    ax.set_ylabel(col2)
    ax.set_zlabel(col3)
    ax.set_title(f'{col1} vs {col2} vs {col3}')

plt.tight_layout()
plt.show()
```

From the diagram above, across the features it would seem that we have somewhat clear separation between the clusters. When we count the number of instances in each cluster, we find that they are somewhat close to the actual counts in the paper (4898 White and 1599 Red):

```python
X2['clusterLabel'] = np.where(X2['cluster'] == 1, 'White', 'Red')
X_standard['clusterLabel'] = np.where(X_standard['cluster'] == 1, 'White', 'Red')

Y2['clusterLabel'] = np.where(Y2['cluster'] == 1, 'White', 'Red')
X2['clusterLabel'].value_counts()
```

The summarized physiochemical data bewteen our clustered datapoints, and the ones for Red and White in the paper, are very close. Therefore, our cluster is a good enough proxy for red and white

# Support Vector Machine using Grid Search over various parameters

**Model:** paper recommends that we use Support Vector Machine to train the model. The paper also suggests that we use RBF kernel (has less parameter), and that SVM on this dataset is sensitive to hyperparameters C (the margin of boundary) and gamma. Here I will tune these two hyperparameters using a 5-fold cross validation grid search

**Metric Evaluation**: Given that from EDA we see an imbalanced class, accuracy may not be the best metric. Instead, I will use **weighted-f1** score as the metric to tune and build the model (weighted because this is a multiclass model). However, I will then look at the confusion metrics and all the relevant metrics (accuracy, f1, recall, precision) to get a full picture of the model performance.


```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score
from sklearn.svm import SVC
from numpy import math
```

```python
class SVMGridSearch:
  """ Handles the tasks of running Grid Search over a parameter grid on an SVM model, retain and plotting results"""

  def __init__(self,X,Y,clabel):
    self.X = X
    self.Y = Y
    self.clabel = clabel
    self.gridSearch = None
    self.searchResults = None
    self.bestParam = None
    self.bestScore = None

  def filterLabel(self):
    """prepares data for training"""
    Xfilter = self.X[self.X['clusterLabel'] == self.clabel].drop(['clusterLabel', 'cluster'], axis = 1)
    Yfilter = self.Y[self.Y['clusterLabel'] == self.clabel].drop(['clusterLabel', 'cluster'], axis = 1)
    return Xfilter, Yfilter

  def runGridSearch(self,paramGrid, scoringMethod):
    """runs the grid search over given parameter grid and scoring method"""
    Xdrop, Ydrop = self.filterLabel()
    self.gridSearch = GridSearchCV(SVC(), paramGrid, cv=5, scoring = scoringMethod)
    self.gridSearch.fit(Xdrop, Ydrop)

    self.searchResults = pd.DataFrame(self.gridSearch.cv_results_)
    self.bestParam = self.gridSearch.best_params_
    self.bestScore = self.gridSearch.best_score_

  def plotHeatMap(self, param1, param2):
    """plot heatmap of gridsearch based on the stated parameters"""
    heatmap_df = self.searchResults.pivot(index=param1, columns=param2, values='mean_test_score')
    plt.figure(figsize=(8, 6))
    sns.heatmap(heatmap_df, annot=True, fmt='.3f', cmap='viridis')
    plt.title(f"Heatmap of F1 Score for {self.clabel} ({param1} vs {param2})")
    plt.xlabel(f'{param2}')
    plt.ylabel(f'{param1}')
    plt.show()

  def plotConfMatrix(self):
    """plot confusion matrix for best model"""
    Xdrop, Ydrop = self.filterLabel()
    Ypred = self.gridSearch.best_estimator_.predict(Xdrop)
    CM = confusion_matrix(Ydrop, Ypred)

    # print out relevant scores
    f1 = f1_score(Ydrop, Ypred, average='weighted')
    print(f"F1 Score: {f1:.2f}")

    precision = precision_score(Ydrop, Ypred, average='weighted')
    print(f"Precision: {precision:.2f}")

    recall = recall_score(Ydrop, Ypred, average='weighted')
    print(f"Recall: {recall:.2f}")

    accuracy = accuracy_score(Ydrop, Ypred)
    print(f"Accuracy: {accuracy:.2f}")

    # plot

    plt.figure(figsize=(6, 4))
    sns.heatmap(CM, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Confusion Matrix for {self.clabel}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

```python
param_grid = {
    'kernel': ['rbf'],
    'C': [0.01, 0.1,1, 10, 50, 100, 500, 1000],
    'gamma': [0.0001, 0.001, 0.01, 0.1, 1]
}

SVMRed = SVMGridSearch(X_standard, Y2, clabel='Red')
SVMRed.runGridSearch(param_grid, 'f1_weighted')

SVMWhite = SVMGridSearch(X_standard, Y2, clabel='White')
SVMWhite.runGridSearch(param_grid, 'f1_weighted')
```

The outputs below show


1. A heatmap of parameter gamma against C, and the resulting f1-weighted scores. We find that the best f1-weighted scores are obtained for Red around C = 10 and gamma = 0.01 regions, and for White around C = 50 and gamma = 0.01 regions

2. Using these hyperparameters, the confusion matrix shows that the final accuracy scores for Red is about 60% for both Red and White wines.




```python
SVMRed.plotHeatMap('param_gamma', 'param_C')
SVMWhite.plotHeatMap('param_gamma', 'param_C')
```

```python
# confusion matrix for best Model
SVMRed.plotConfMatrix()
SVMWhite.plotConfMatrix()
```
